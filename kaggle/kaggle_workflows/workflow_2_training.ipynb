# Workflow 2: Training

This notebook is designed to facilitate the training of the Persona-Consistent Chatbot model using the specified configurations and datasets. It includes steps for loading data, initializing the model, and executing the training process.

```python
# Import necessary libraries
import os
import yaml
from src.training.sft_trainer import SFTTrainer
from src.utils.config_loader import load_config

# Load configuration files
base_config = load_config('../configs/base_config.yaml')
data_config = load_config('../configs/data_config.yaml')
model_config = load_config('../configs/model_config.yaml')
sft_config = load_config('../configs/sft_config.yaml')

# Set up directories
os.makedirs(data_config['processed_data_dir'], exist_ok=True)
os.makedirs(model_config['model_save_dir'], exist_ok=True)

# Initialize the trainer
trainer = SFTTrainer(base_config, data_config, model_config, sft_config)

# Start training
trainer.train()
```
